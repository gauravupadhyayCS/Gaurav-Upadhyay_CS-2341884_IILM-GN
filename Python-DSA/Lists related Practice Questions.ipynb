```python
'''Program to increment the elements of a list by 5. The list is passed as an argument to a function'''
def increment_list(lst):
    for i in range(len(lst)):
        lst[i] = lst[i] + 5


numbers = list(map(int, input("Enter list elements: ").split()))
increment_list(numbers)
print("Updated list:", numbers)
```

```python
#Write a program to find the number of times an element occurs in the list.
numbers=list(map(int,input("enter the numbers:").split()))
elements=int(input("enter the element:"))
count=0
for i in numbers:
    if i==elements:
        count+=1
print("count:",count)
```

```python
''' Write a program to read a list of n integers (positive as well as negative).
Create two new lists, one having all positive numbers and the other having all 
negative numbers from the given list. Print all three lists. '''

numbers=list(map(int,input("enter n integers:").split()))
neg_int=[]
pos_int=[]
for i in numbers:
    if i>0:
        pos_int.append(i)
    else:
        neg_int.append(i)
print("Original list:",numbers)
print("Positive numbers:",pos_int)
print("Negative numbers:",neg_int)
```
