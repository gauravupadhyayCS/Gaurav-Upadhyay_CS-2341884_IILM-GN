```python
class Node:
    def __init__(self, data = 0, next = None):
        self.data = data
        self.next = next
```

```python

a = Node(13)
b = Node(14)
a.next = b
c = Node(30)
b.next = c
d = Node(40)
e = Node(50)
c.next = d
d.next = e
```

```python

# Function to print the LL
def printLL(head):
    while head is not None:
        print(str(head.data) + ' -> ', end ='')
        head = head.next
    print(None)
    return
printLL(a)
```

```python
def takeInput():
        # taking the space seperated input in a list
        inputList = [int(ele) for ele in input().split()]
        head = None
        #Traversing the List ending with -1
        for currData in inputList:
            if currData == -1:
                break
            # Creating a new Node and checking is it's head, if not, traversing the list to find the end and appending element
            newNode = Node(currData)
            if head is None:
                head = newNode
            else:
                curr = head
                while curr.next is not None:
                    curr.next = newNode
        return head
    
def printLL(head):
    while head is not None:
        print(str(head.data) + '->', end = '')
        head = head.next
    print('None')
    return
```

```python

head = takeInput()
printLL(head)
```
