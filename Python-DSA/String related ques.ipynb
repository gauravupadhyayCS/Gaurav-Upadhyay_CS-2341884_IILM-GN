```python
#1. Write a program with a user defined function to count the number of times a character (passed as argument) occurs in the given string.
def count_char(s,ch):
    count=0
    for i in s:
        if i==ch:
            count+=1
    return count

s=input("enter the string:")
ch=input("enter the character to be counted:")

result= count_char(s,ch)
print("count:",result)
```

```python
'''Write a program with a user defined function with string 
as a parameter which replaces all vowels in the string with '*'.'''
def repl_vowel(s):
    res=''
    for ch in s:
        if ch.lower() in "aeiou":
            res+='*'
        else:
            res+=ch
    return res
s=input("enter the string:")
result= repl_vowel(s)
print(result)
            
```

```python
''' Write a program which reverses a string passed as parameter
and stores the reversed string in a new string. 
Use a user defined function for reversing the string.'''

def rever_str(s):
    rev=''
    for ch in s:
        rev=ch + rev
    return rev

s=input("enter the string:")
result= rever_str(s)
print(result)
```

```python
'''Write a program using a user defined function to check if a string is a palindrome or not. 
(A string is called palindrome if it reads same backwards as forward. For example, Kanak is a palindrome.)'''

def palin(s):
    rev=''
    for ch in s:
        rev= ch + rev
    if s==rev:
        return True
    else:
        return False

s=input("enter the string:")
if palin(s):
    print("palindrome")
else:
    print("not palin")
```

```python
'''  Write a user defined function to convert a string with more than one word into 
title case string where string is passed as parameter. 
(Title case means that the first letter of each word is capitalised).'''

def title(s):
    words=s.split()
    res=''
    for word in words:
        res+=word[0].upper() + word[1:] + " " 
    return res.strip()
s=input("enter the string:")
print("new string:",title(s))
```

```python
''' Write a function deleteChar() which takes two parameters one is a string and other is a character. 
The function should create a new string after deleting all occurrences of the character from 
the string and return the new string. '''

def deletechar(s,ch):
    res=''
    for c in s:
        if c!=ch:
            res+=c
    return res
s=input("enter the string:")
ch=input("enter the character to be removed:")
print(" new string ",deletechar(s,ch))
```

```python
''' Input a string having some digits. 
Write a function to return the sum of digits present in this string. '''

def sumofdigits(s):
    total=0
    for ch in s:
        if ch.isdigit():
            total+=int(ch)
    return total
s=input("enter the string:")
print("sum of digits in string :",sumofdigits(s))
            
```
